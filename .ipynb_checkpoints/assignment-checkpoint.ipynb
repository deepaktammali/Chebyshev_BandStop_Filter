```python
import scipy.signal as signal
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Required Digital Frequency Response
```

```python
sf1a = 2*np.tan((np.pi/4)/2)
```

```python
sf2a = 2*np.tan((np.pi/3)/2)
```

```python
pf1a = 2*np.tan((np.pi/6)/2)
```

```python
pf2a = 2*np.tan((2*(np.pi)/3)/2)
```

```python
pf1a,sf1a,sf2a,pf2a  #Passbandandstopbandangularfrequencies
```

```python
x_axis=np.arange(0,np.pi,step=0.01,dtype=float)
```

```python
def conversion(element):
    return 1 if (element>=0 and element<=np.pi/6 or element>=np.pi*(2/3) and element<=np.pi) else 0
```

```python
y_axis =list(map(conversion,x_axis))
```

```python
plt.plot(x_axis,y_axis)
plt.tight_layout()
plt.ylabel("Magnitude")
plt.xlabel("Digital Frequency")
plt.grid()
plt.title("Ideal Filter Frequency Response")
```

```python
gpass = -20*np.log10(0.707)
gpass
```

```python
gstop = -20*np.log10(0.3)
gstop
```

```python
N, Wn = signal.cheb1ord(wp = [pf1a,pf2a], ws = [sf1a,sf2a],analog=True,gpass=gpass,gstop=gstop)
```

```python
print(f"Order of chebyshev filter is : {ordcheb}")
```

```python
print(f"The Three Db Frequency is : {Wn[0]} radians")
```

```python
Wn
```

```python
b,a=signal.cheby1(N, gpass,Wn, btype='bandstop', analog=True, output='ba')
```

```python
Numerator=""
power=0
for coefficient in b:
    Numerator = Numerator + f"{round(coefficient,3)}Z^{power}"+" + "
    power=power+1
Numerator = Numerator[:-3]
print(f"Numerator: {Numerator} ")
```

```python
denominator=""
power=0
for coefficient in a:
    denominator = denominator + f"{round(coefficient,3)}Z^{power}"+" + "
    power=power+1
denominator = denominator[:-3]
print(f"Denominator: {denominator} ")
```

```python
w,h = signal.freqs(b, a)
```

```python
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Chebyshev Type I frequency response (rp=3db)') #ripple 3db
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.axvline(Wn[0], color='green') # cutoff frequency
plt.axhline(-3, color='green') # rp
plt.show()
```

```python
znum,pdem = signal.bilinear(b, a, fs=1.0)
```

```python
Numeratorb=""
power=0
for coefficient in znum:
    Numeratorb = Numeratorb + f"{round(coefficient,3)}Z^{power}"+" + "
    power=power+1
Numeratorb = Numeratorb[:-3]
print(f"Numerator: {Numeratorb} ")
```

```python
denominatorb=""
power=0
for coefficient in pdem:
    denominatorb = denominatorb + f"{round(coefficient,3)}Z^{power}"+" + "
    power=power+1
denominatorb = denominatorb[:-3]
print(f"Denominator: {denominatorb} ")
```

```python
wdig,hdig = signal.freqz(znum,pdem,worN=1024)
```

```python
plt.plot(wdig, 20 * np.log10(abs(hdig)), 'b')
plt.title('Digital filter frequency response')
plt.xlabel('Angle (radians)', color='b')
plt.ylabel('Amplitude [dB]', color='b')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.plot(wdig, 20 * np.log10(abs(hdig)), 'b')
plt.fill_between(wdig[0:175],-50,20 * np.log10(abs(hdig))[0:175],color='grey')
plt.fill_between(wdig[678:],-50,20 * np.log10(abs(hdig))[678:],color='grey')
plt.fill_between(wdig[260:340],-50,20 * np.log10(abs(hdig))[260:340],color='grey')

plt.fill_between(wdig[175:260],-50,20 * np.log10(abs(hdig))[175:260],color='orange')
plt.fill_between(wdig[340:678],-50,20 * np.log10(abs(hdig))[340:678],color='orange')

plt.title('Digital filter frequency response')
plt.xlabel('Angle (radians)', color='b')
plt.ylabel('Amplitude [dB]', color='b')
plt.axvline(np.pi/6, color='green') # passband
plt.axvline(np.pi*(2/3), color='green') # Passband
plt.axvline(np.pi/3, color='orange') # StropBand
plt.axvline(np.pi/4, color='orange') # StopBand
plt.axhline(-3, color='black') # rp
plt.axhline(-10.4, color='black') # rp
plt.grid()
plt.tight_layout()
plt.show()


#GREY represents Pass Band and Stop Band
#Orange represents Transition Band
```

```python
import librosa as lr
```

```python
audio , sfreq = lr.load('sound1.wav')
```

```python
print(f"Number of Samples in the audio signal : {len(audio)}")
```

```python
print(f"Sampling Frequency : {sfreq}")
```

```python
#Generating Time Stamps
```

```python
time_range = np.arange(0,len(audio))/sfreq
```

```python
plt.plot(time_range,(audio+0.6))
plt.title("Input Audio Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
plt.plot(time_range[:100],(audio+0.6)[:100])
plt.title("A Part of the Input Audio Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
yf = scipy.fft(audio)
plt.plot(time_range[:110250], yf[:110250])
plt.title("Discrete Fourier Tranform of Input Audio Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
yf = scipy.fft(audio)
plt.plot(time_range[:100], yf[:100])
plt.title("A Part of Discrete Fourier Tranform of Input Audio Signal")
plt.xlabel("Discrete Frequency")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
plt.plot(time_range[:100],(filteredoutput+0.6)[:100])
plt.title("A Part of the Filtered Input Audio Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```
